```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data=pd.read_csv('/content/drive/MyDrive/50_Startups.csv')
data
```

```python
data = pd.get_dummies(data, columns=["State"], drop_first=True)
```

```python
x=data.drop("Profit",axis=1).values
y=data["Profit"].values
```

```python
y.reshape(-1,1).shape
```

**Logistic Regression model**

```python
t = np.median(y)
y_new = np.where(y > t, 1, 0)

model = LogisticRegression(C=0.5)
model.fit(x, y_new)
```

```python
y_pred = model.predict(x)
print(y_pred)
```

```python
print(classification_report(y_new, y_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
import numpy as np

# Assuming 'x' and 'y' are your features and target variable
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Convert continuous target to discrete classes using median as threshold
t = np.median(y_train)  # Calculate median on training data
y_train_new = np.where(y_train > t, 1, 0) # Create binary labels for training
y_test_new = np.where(y_test > t, 1, 0) # Create binary labels for testing


nb = GaussianNB()
nb.fit(X_train, y_train_new) # Fit with discrete class labels

y_pred_nb = nb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test_new, y_pred_nb)  # Compare predicted to true binary labels
print(accuracy)

print(classification_report(y_test_new, y_pred_nb))
```
