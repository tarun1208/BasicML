The necessary libraries are imported such as Matplotlib, Seaborn, Pandas, Numpy

````python
#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
````

````python
#Reading or loading data
df = pd.read_csv('/content/drive/MyDrive/ICRISAT-District Level Data.csv')
````

````python
#Dropping unneccesary columns
df = df.drop(columns=[col for col in df.columns if "Negative" in col],axis=1)
````

The data has been imported as a dataset into variable ```data``` using the method ```.read_csv()``` method of pandas library


````python
#check first five rows
df.head()

#check missing values
df.isnull().sum()

#Statistics of numeric columns
df.describe()

#check data type of columns
df.dtypes
````

````python
# Plot state-wise frequency
plt.figure(figsize=(12, 6))
state_counts = df['State Name'].value_counts()
sns.barplot(x=state_counts.index, y=state_counts.values, palette='viridis')
plt.xticks(rotation=90)
plt.title('State Frequency')
plt.xlabel('State Name')
plt.ylabel('Frequency')
plt.show()
````

The above data contains statewise Agriculture details and Dropping unnecessary columns and representing sate frequency

````python
df= df.drop_duplicates()
df.count()
````

**Distribution of Area and Production for Different Crops**

````python
# Plot histograms for area and production of selected crops
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.histplot(df['RICE AREA (1000 ha)'], kde=True, ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Distribution of Rice Area')

sns.histplot(df['RICE PRODUCTION (1000 tons)'], kde=True, ax=axes[0, 1], color='green')
axes[0, 1].set_title('Distribution of Rice Production')

sns.histplot(df['WHEAT AREA (1000 ha)'], kde=True, ax=axes[1, 0], color='red')
axes[1, 0].set_title('Distribution of Wheat Area')

sns.histplot(df['WHEAT PRODUCTION (1000 tons)'], kde=True, ax=axes[1, 1], color='orange')
axes[1, 1].set_title('Distribution of Wheat Production')

plt.tight_layout()
plt.show()
````

````python
#plotting heatmap for missing values
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values')
plt.show()
````

**Correlation Heatmap**

````python
# Select relevant columns for correlation
corr_columns = ['RICE YIELD (Kg per ha)', 'WHEAT YIELD (Kg per ha)', 'MAIZE YIELD (Kg per ha)', 'BARLEY YIELD (Kg per ha)', 'COTTON YIELD (Kg per ha)']
corr_df = df[corr_columns]

# Plot correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap between Crop Yields')
plt.show()
````

````python
# Create a boxplot for crop yields
plt.figure(figsize=(10, 6))
yield_df = df[['RICE YIELD (Kg per ha)', 'WHEAT YIELD (Kg per ha)', 'MAIZE YIELD (Kg per ha)', 'BARLEY YIELD (Kg per ha)', 'COTTON YIELD (Kg per ha)']]
sns.boxplot(data=yield_df)
plt.title('Boxplot of Crop Yields')
plt.ylabel('Yield (Kg per ha)')
plt.xticks(rotation=90)
plt.show()
````

````python
# Scatter plots for area vs production
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.scatterplot(data=df, x='RICE AREA (1000 ha)', y='RICE PRODUCTION (1000 tons)', ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Rice Area vs Rice Production')

sns.scatterplot(data=df, x='WHEAT AREA (1000 ha)', y='WHEAT PRODUCTION (1000 tons)', ax=axes[0, 1], color='green')
axes[0, 1].set_title('Wheat Area vs Wheat Production')

sns.scatterplot(data=df, x='MAIZE AREA (1000 ha)', y='MAIZE PRODUCTION (1000 tons)', ax=axes[1, 0], color='red')
axes[1, 0].set_title('Maize Area vs Maize Production')

sns.scatterplot(data=df, x='COTTON AREA (1000 ha)', y='COTTON PRODUCTION (1000 tons)', ax=axes[1, 1], color='orange')
axes[1, 1].set_title('Cotton Area vs Cotton Production')

plt.tight_layout()
plt.show()
````

**Year-Wise Crop Data**

````python
# Line plot to show trend over years for a specific crop (e.g., Rice)
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='RICE PRODUCTION (1000 tons)', hue='State Name')
plt.title('Year-wise Rice Production Trend')
plt.xlabel('Year')
plt.ylabel('Rice Production (1000 tons)')
plt.legend(title='State Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
````

**Distribution of Total Area and Total Production**

````python
# Sum total area and total production
df['Total Area'] = df[['RICE AREA (1000 ha)', 'WHEAT AREA (1000 ha)', 'MAIZE AREA (1000 ha)', 'COTTON AREA (1000 ha)']].sum(axis=1)
df['Total Production'] = df[['RICE PRODUCTION (1000 tons)', 'WHEAT PRODUCTION (1000 tons)', 'MAIZE PRODUCTION (1000 tons)', 'COTTON PRODUCTION (1000 tons)']].sum(axis=1)

# Plot the total area and production
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Total Area', y='Total Production', color='purple')
plt.title('Total Area vs Total Production')
plt.xlabel('Total Area (1000 ha)')
plt.ylabel('Total Production (1000 tons)')
plt.show()
````
